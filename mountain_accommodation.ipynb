```python
from requests import get
from bs4 import BeautifulSoup
from time import sleep             #不能狂爬,要用 sleep 設停頓避免被反爬蟲禁IP
import openpyxl
import pygsheets
import pandas as pd

# 抓金鑰
gc = pygsheets.authorize(service_file='ccclub_api_key.json')
# 連接雲端 sheet
sht = gc.open_by_url(
'https://docs.google.com/spreadsheets/d/1t7U4pvptpD_cfTMD4tUwDeCffAopPQF2hJEt386J18A/'
)


dic1 = {}
info_all = []
url_base = "https://hiking.biji.co"
page = 0
url_house = "https://hiking.biji.co/index.php?q=chalet&keyword=&page="
data=["init"]
```

```python
worksheet_title = 'chalet'  
ws = sht.worksheet_by_title(worksheet_title)
df = pd.DataFrame()
```

```python
while len(data) != 0:
    page += 1
    res = get(url_house+str(page))
    soup = BeautifulSoup(res.text)
    data = soup.find('div', class_="bg-white rounded overflow-hidden shadow-z1").find_all('li', class_='flex')
    for i in range(len(data)):
        name = data[i].a["title"]
        url = url_base + data[i].a["href"]
        sub_res = get(url)
        sub_soup = BeautifulSoup(sub_res.text).find('dl', class_="border")
        sub_data_tag =sub_soup.find_all('dt')
        sub_data_value =sub_soup.find_all('dd')
        info = [""]*100
        for j in range(len(sub_data_tag)):
            if sub_data_tag[j].text in ["Google Map","Open Street Map"]:
                pass
            elif sub_data_tag[j].text == "所屬步道":
                link = sub_data_value[j].find_all('li')
                linklst = []
                for k in range(len(link)):
                    if link[k].a == None:
                        linklst.append(link[k].text.strip(" "))
                    else:
                        linklst.append(link[k].a['title'] + " " + (url_base+link[k].a['href']))
                link_info = "\n".join(linklst)
                dic1[sub_data_tag[j].text] = dic1.get(sub_data_tag[j].text,len(dic1))
                info[dic1[sub_data_tag[j].text]] = link_info
            else:
                dic1[sub_data_tag[j].text] = dic1.get(sub_data_tag[j].text,len(dic1))
                info[dic1[sub_data_tag[j].text]] = sub_data_value[j].text
        info_all.append([name]+info)
```

```python
title = ["山屋名稱"]+list(dic1.keys())
title += [""]*(101-len(title))
df = pd.DataFrame(columns=title)
#ws.append(["山屋名稱"]+list(dic1.keys()))
for i in range(len(info_all)):
    df = df.append(pd.Series(info_all[i], index=df.columns), ignore_index=True)
ws.set_dataframe(df,(0,0))
```
